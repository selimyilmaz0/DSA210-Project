```python
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv("data/master_dataset_1980_2024.csv")


avg_migrant_percent = df.groupby("Year")["Migrant % of Population"].mean()

# Plot
plt.figure(figsize=(12, 6))
plt.plot(avg_migrant_percent.index, avg_migrant_percent.values, marker='o')
plt.title("Average Migrant % of Population Over Time (19802024)")
plt.xlabel("Year")
plt.ylabel("Migrant % of Population")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Migrant % vs GDP

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Migrant % of Population", y="GDP", hue="Country Name", alpha=0.7)
plt.title("Migrant % of Population vs GDP (19802024)")
plt.xlabel("Migrant % of Population")
plt.ylabel("GDP (Current US$)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Migrant % vs GDP Correlarion

```python
from scipy.stats import pearsonr


subset = df[["Migrant % of Population", "GDP"]].dropna()

# Pearson Correlation
corr_coef, p_value = pearsonr(subset["Migrant % of Population"], subset["GDP"])

print(f"Correlation coefficient (Migrant % vs GDP): {corr_coef:.3f}")
print(f"P-value: {p_value:.3e}")
```
